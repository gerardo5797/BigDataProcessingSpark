```
### CLEANING
```

```
## 1. Get rid of dropoff > pickup (GREEN)

df_green_clean = df_green.where(df_green.lpep_pickup_datetime < df_green.lpep_dropoff_datetime)
df_green_clean.count()
```

```
## 1. Get rid of dropoff > pickup (YELLOW)
df_yellow_clean = df_yellow.where(df_yellow.tpep_pickup_datetime < df_yellow.tpep_dropoff_datetime)
df_yellow_clean.count()
```

```
# 2. Get rid of Trips with negative speed (Green)
# Calculate trip duration in hours (Green)

df_green_clean = df_green_clean.withColumn('Duration',(F.col("lpep_dropoff_datetime").cast("long") - F.col('lpep_pickup_datetime').cast("long"))/3600)
```

```
# 2. Get rid of Trips with negative speed (Green)
# Calculate Speed in mph (Green)

df_green_clean = df_green_clean.withColumn('Speed',(F.col("trip_distance") / F.col('Duration')))
```

```
# 2. Get rid of Trips with negative speed (Green)

df_green_clean = df_green_clean.where(df_green_clean.Speed >= 0)
df_green_clean.count()
```

```
# 2. Get rid of Trips with negative speed (Yellow)
# Calculate trip duration in hours (Yellow)

df_yellow_clean = df_yellow_clean.withColumn('Duration',(F.col("tpep_dropoff_datetime").cast("long") - F.col('tpep_pickup_datetime').cast("long"))/3600)
```

```
# 2. Get rid of Trips with negative speed (Yellow)
# Calculate Speed in mph (Yellow)

df_yellow_clean = df_yellow_clean.withColumn('Speed',(F.col("trip_distance") / F.col('Duration')))
```

```
# 2. Get rid of Trips with negative speed (Yellow)

df_yellow_clean = df_yellow_clean.where(df_yellow_clean.Speed >= 0)
df_yellow_clean.count()
```

```
#3. Get rid of Trips with very high speed, 50 mph is the limit in all counties, will have a 10% margin, use 55 mph) (GREEN)

df_green_clean = df_green_clean.where(df_green_clean.Speed <= 55)
df_green_clean.count()
```

```
#3. Get rid of Trips with very high speed, 50 mph is the limit in all counties, will have a 10% margin, use 55 mph) (YELLOW)

df_yellow_clean = df_yellow_clean.where(df_yellow_clean.Speed <= 55)
df_yellow_clean.count()
```

```
# 4. Trips that are travelling too short duration wise (less than 1 min) (GREEN)

df_green_clean = df_green_clean.where(df_green_clean.Duration >=0.016)
df_green_clean.count()
```

```
# 4. Trips that are travelling too short duration wise(less than 1 min) (YELLOW)

df_yellow_clean = df_yellow_clean.where(df_yellow_clean.Duration >=0.016)
df_yellow_clean.count()
```

```
# 5. Trips that are travelling too long duration wise (more than 2h) (GREEN)

df_green_clean = df_green_clean.where(df_green_clean.Duration <=2)
df_green_clean.count()
```

```
# 5. Trips that are travelling too long duration wise (more than 2h) (YELLOW)

df_yellow_clean = df_yellow_clean.where(df_yellow_clean.Duration <=2)
df_yellow_clean.count()
```

```
# 6. Trips that are travelling too short distance wise (less than 2 miles) (GREEN)
df_green_clean = df_green_clean.where(df_green_clean.trip_distance >=0.2)
df_green_clean.count()
```

```
# 6. Trips that are travelling too short distance wise (less than 2 miles) (YELLOW)
df_yellow_clean = df_yellow_clean.where(df_yellow_clean.trip_distance >=0.2)
df_yellow_clean.count()
```

```
#7. Trips that are travelling too long (distance wise, more than 50 miles) GREEN

df_green_clean = df_green_clean.where(df_green_clean.trip_distance <=50)
df_green_clean.count()
```

```
#7. Trips that are travelling too long (distance wise, more than 50 miles) GREEN

df_yellow_clean = df_yellow_clean.where(df_yellow_clean.trip_distance <=50)
df_yellow_clean.count()
```

```
#8. Trips that finished before First of January 2019. (GREEN)
df_green_clean = df_green_clean.where(df_green_clean.lpep_dropoff_datetime >= '2019-01-01T00:00:00.000+0000')
df_green_clean.count()
```

```
#8. Trips that finished before First of January 2019. (YELLOW)
df_yellow_clean = df_yellow_clean.where(df_yellow_clean.tpep_dropoff_datetime >= '2019-01-01T00:00:00.000+0000')
df_yellow_clean.count()
```

```
#8. Trips that finished after 30th of April 2022. (GREEN)
df_green_clean = df_green_clean.where(df_green_clean.lpep_dropoff_datetime <= '2022-04-30T23:59:59.000+0000')
df_green_clean.count()
```

```
#8. Trips that finished after 30th of April 2022. (YELLOW)
df_yellow_clean = df_yellow_clean.where(df_yellow_clean.tpep_dropoff_datetime <= '2022-04-30T23:59:59.000+0000')
df_yellow_clean.count()
```

```
# 9. Add Taxi Colour Field (GREEN)

df_green_clean = df_green_clean.withColumn("taxi_colour", lit('Green'))
```

```
# 9. Add Taxi Colour Field (YELLOW)

df_yellow_clean = df_yellow_clean.withColumn("taxi_colour", lit('Yellow'))
```
