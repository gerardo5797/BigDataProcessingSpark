```
### Create summary table to later join with top days and hours
spark.sql('''
CREATE OR REPLACE TABLE sum_table
SELECT YEAR_MONTH,
count(*) as number_trips,
round(avg(passenger_count),2) as avg_pass,
round(avg(total_amount),2) as avg_amount_paid,
round((sum(total_amount)/ sum(passenger_count)),2) as avg_amount_per_pass
from taxi_final
group by YEAR_MONTH
''')
```

```
# Check output for summary table
spark.sql('''
SELECT YEAR_MONTH,
count(*) as number_trips,
round(avg(passenger_count),2) as avg_pass,
round(avg(total_amount),2) as avg_amount_paid,
round((sum(total_amount)/ sum(passenger_count)),2) as avg_amount_per_pass
from taxi_final
group by YEAR_MONTH
''').display()
```

```
# Create weekday table (will extract top day later)
spark.sql('''
CREATE OR REPLACE TABLE weekday_table
SELECT
year_month,
week_day,
count(*) as number_of_trips
from taxi_final
group by year_month, week_day
order by year_month asc , number_of_trips desc
''')
```

```
# Check output for weekday table
spark.sql('''
SELECT
year_month,
week_day,
count(*) as number_of_trips
from taxi_final
group by year_month, week_day
order by year_month asc , number_of_trips desc
''').display()
```

```
# Assign ranking to weekday table, create day_ranked table
spark.sql('''
CREATE OR REPLACE TABLE day_ranked
SELECT year_month,week_day, number_of_trips,
RANK() OVER (PARTITION BY year_month ORDER BY number_of_trips desc) AS RK 
from weekday_table
''')
```

```
# See output day_ranked
spark.sql('''
SELECT year_month,week_day, number_of_trips,
RANK() OVER (PARTITION BY year_month ORDER BY number_of_trips desc) AS RK 
from weekday_table
''').display()
```

```
spark.sql('''
CREATE OR REPLACE TABLE top_weekday
select year_month, week_day
from day_ranked
where RK=1
''')
```

```
# See output top_weekday
spark.sql('''
select year_month, week_day
from day_ranked
where RK=1
''').display()
```

```
# Create Hours Table
spark.sql('''
CREATE OR REPLACE TABLE hours_table
SELECT
year_month,
trip_hour,
count(*) as number_trips
from taxi_final
group by year_month, trip_hour
order by year_month asc , number_trips desc
''')
```

```
# Create table hours_rank
spark.sql('''
CREATE OR REPLACE TABLE hours_rank
SELECT year_month,trip_hour, number_trips,
RANK() OVER (PARTITION BY year_month ORDER BY number_trips desc) AS RK 
from hours_table
''')
```

```
# See Hours Rank
spark.sql('''
SELECT year_month,trip_hour, number_trips,
RANK() OVER (PARTITION BY year_month ORDER BY number_trips desc) AS RK 
from hours_table
''').display()
```

```
# Create top_hours table
spark.sql('''
create OR REPLACE table top_hours
select year_month, trip_hour
from hours_rank
where RK=1
''')
```

```
# See top_hours table
spark.sql('''
select year_month, trip_hour
from hours_rank
where RK=1
''').display()
```

```
# Query to join the 3 tables and get the requested output
spark.sql('''
SELECT top_hours.year_month, top_hours.trip_hour, top_weekday.week_day, sum_table.number_trips, sum_table.avg_pass, sum_table.avg_amount_paid, sum_table.avg_amount_per_pass
FROM top_hours
LEFT JOIN top_weekday
ON top_hours.year_month = top_weekday.year_month
LEFT JOIN sum_table
ON top_hours.year_month = sum_table.year_month
''').display()


```

```
#Question 5: Add duration bin column
spark.sql('''
ALTER TABLE taxi_final ADD column (duration_bin string)
''')
```

```
#Question 5: Add duration bin column
spark.sql('''
UPDATE taxi_final SET duration_bin =
(CASE WHEN Duration BETWEEN 0/60 AND 5/60 THEN '0-5'
WHEN Duration BETWEEN 5.0000001/60 AND 10/60 THEN '5-10'
WHEN Duration BETWEEN 10.0000001/60 AND 20/60 THEN '10-20'
WHEN Duration BETWEEN 20.0000001/60 AND 30/60 THEN '20-30'
WHEN Duration BETWEEN 30.0000001/60 AND 60/60 THEN '30-60'
WHEN Duration BETWEEN 60.0000001/60 AND 10000/60 THEN 'over60'
ELSE 'unknown'
END)
''')
```

```
#Question 5: Query
spark.sql('''
select duration_bin,
round((avg(Speed)*1.60934),2) as avg_speed,
round((avg((trip_distance*1.60934) / total_amount)),2) as avg_km_per_dollar
from taxi_final
group by duration_bin
order by duration_bin asc
''').display()
```

```
#Question 6: Query
spark.sql('''
select duration_bin,
round((avg(Speed)*1.60934),2) as avg_speed,
round((avg((trip_distance*1.60934) / total_amount)),2) as avg_km_per_dollar,
round((avg((Duration*60) / total_amount)),2) as avg_min_per_dollar,
count (*) as number_of_trips,
round(sum(total_amount),2) as total_amount,
round(avg(total_amount),2) as avg_amount
from taxi_final
group by duration_bin
order by duration_bin asc
''').display()
```
